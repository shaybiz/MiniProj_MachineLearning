```python
%matplotlib inline 
import matplotlib.pylab as plt
plt.rcParams['figure.figsize'] = (12,6)
import numpy as np
import pandas as pd
```

**The special thing about time series is data points relate to one another. They are not independent. Expanding window takes all data into consideration while rolling window relates to recent data only.**

```python
# Window functions are like aggregation functions
# You can use them in conjunction with .resample() which will soon be a deferred operation
```

```python
df = pd.DataFrame(np.random.randn(600, 3), index = pd.date_range('7/1/2016', freq = 'S', periods = 600), columns = ['A', 'B', 'C'])
```

```python
#pd.rolling_mean(df, window = 2)[1:10] # in future versions you want to resample separately
r = df.rolling(window = 10)
#r.agg, r.apply, r.count, r.exclusions, r.max, r.median, r.name, r.quantile, r.kurt, r.cov, r.corr, r.aggregate, r.std, r.skew, r.sum, r.var
#df.plot(style = 'k--')
df.plot()
#r.mean().plot(style = 'r')
r.mean().plot()
```

```python
df['A'].plot(color = 'gray')
r.mean()['A'].plot(color = 'red')
```

```python
# can also easily plot a rolling average per column
df = pd.DataFrame(np.random.randn(1000, 4), index = pd.date_range('6/6/16', periods = 1000), columns = ['A', 'B', 'C', 'D'])
```

```python
df.head()
```

```python
df = df.cumsum()
df.rolling(window = 50).sum().plot(subplots=True)
```

```python
# What about a custom function?
df.rolling(window = 10).apply(lambda x: np.fabs(x - x.mean()).mean())
```

```python
# Expanding windows
# Yields the value of the statistic with all the data available up to that point in time
df.expanding(min_periods = 1).mean()[1:5]
```

```python
df.expanding(min_periods = 1).mean().plot()
#expanding windows are spiky at the beginning when the window is small, but level off as more and more data is added
```

```python
# Will an expanding window provide a more or less stable statistic than a rolling statistic? 
# When might one vs. the other be appropriate?
```

**Expanding window use ALL data available upto a given point to compute the statistic (the window is not fixed; it keeps expanding with each addition of data), and hence is it more stable that rolling statistics. Expanding window is highly dependent on sample size. The rolling window is better in computing short period changes. Exapnding window is useful when old data is as relevant as new additions to data.**

```python
# EWMA
# Why use EWMA?
```

**We use EWMA if we want to give more weight to recent data. This enables a fast adaptation to any change in data.**

```python
# Go do it!
# pandas.pydata.org/pandas-docs/stable/computation.html#stats-aggregate
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                          index=pd.date_range('1/1/2000', periods=1000),
                          columns=['A', 'B', 'C', 'D'])
```

```python
df.rolling(window=len(df), min_periods=1).mean().plot()
df.expanding(min_periods=1).mean().plot()
```

```python
df.rolling(window=int(0.2*len(df)), min_periods=1).mean().plot()
df.expanding(min_periods=1).mean().plot()
```

```python
ts = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/16', freq='D', periods=1000))
ts.ewm(span=60, freq='D', min_periods=0, adjust=True).mean().plot()
ts.rolling(window=60).mean().plot()
```

```python
ts = pd.Series(np.random.randn(20), pd.date_range('7/1/16', freq='D', periods=20))
ts_lagged = ts.shift()
```

```python
plt.plot(ts,color='blue')
plt.plot(ts_lagged,color='red')
```

```python
ts = pd.Series(np.random.randn(20), pd.date_range('7/1/16', freq='H', periods=20))
ts_lagged = ts.shift(5)
```

```python
plt.plot(ts,color='blue')
plt.plot(ts_lagged,color='red')
```

```python
#Change the shift to past instead of to the future.
#Simply change shift argument to negative
ts = pd.Series(np.random.randn(20), pd.date_range('7/1/16', freq='H', periods=20))
ts_lagged = ts.shift(-5)
plt.plot(ts,color='blue')
plt.plot(ts_lagged,color='red')
```
